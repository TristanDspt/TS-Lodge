```python
import pandas as pd
import numpy as np
from tools import sherlock
```

```python
# 2024
df24 = pd.read_csv(r"..\data\raw\2BLONDEL_2024.csv", sep=';', encoding='latin_1', decimal=',')
df24.columns = df24.columns.str.strip().str.replace(' ', '_')
df24['Date_écriture'] = pd.to_datetime(df24['Date_écriture'], format="%d/%m/%Y")

df24 = df24.drop(columns=['Débit_origine', 'Crédit_origine', 'Lettrage_partiel', 'Monnaie', 'ISO_Monnaie', 'taux_change', 'Lettrage', 'Lettrage_n+1'])
df24 = df24.rename(str.lower, axis='columns')
df24['année'] = 2024
```

```python
# 2025
df25 = pd.read_csv(r"..\data\raw\2BLONDEL_2025.csv", sep=';', encoding='cp850', decimal=',')
df25.columns = df25.columns.str.strip().str.replace(' ', '_')
df25['Date_écriture'] = pd.to_datetime(df25['Date_écriture'], format="%d/%m/%Y")

df25 = df25.drop(columns=
    ['Débit_origine', 'Crédit_origine', 'Lettrage_partiel', 'Monnaie', 'ISO_Monnaie', 'Taux_change', 'Lettrage_N', 'Lettrage_N+1',
    'Révision', 'Mois', 'Jour', 'Type_règlement', 'Quantité_1', 'Unité_1', 'Quantité_2', 'Unité_2', 'Unnamed:_25']
    )
df25 = df25.rename(columns={'Compte':'Code_Compte', 'Libellé_compte':'Intitulé_compte', 'Journal':'Code_journal', 'Lettrage_N':'Lettrage'})
df25 = df25.rename(str.lower, axis='columns')
```

# Merge

```python
df = pd.concat([df24, df25])
```

```python
sherlock(df)
```

```python
df.to_csv(r"..\data\processed\2BLONDEL_consolide.csv", index=False, sep=';', decimal=',', encoding='utf-8-sig')
```
