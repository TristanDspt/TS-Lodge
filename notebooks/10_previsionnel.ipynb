# Prévisionnel CA — Cible S1 (US3)

Ce notebook produit une estimation de CA cible à destination du repreneur et du cabinet Axescibles. Il constitue la matière première chiffrée pour l'estimation du bien — la valorisation reste du ressort d'Axescibles.

**Limite :** l'historique disponible (6 mois actifs 2024 + 1 an 2025) ne permet pas de projeter une trajectoire pluriannuelle fiable. On retient donc une cible unique en scénario réaliste LMNP maintenu.

## Modèle

    CA = Jours disponibles × Taux d'occupation × ADR

```python
# =====================
# DONNÉES RÉELLES
# =====================

# --- Depuis taxe de séjour (notebook 7) ---
adr_2024 = 98.28              # €
tx_occ_2024 = 0.503           # %
revpar_2024 = 49.43           # €

# --- Depuis comptabilité (notebook 4) ---
ca_2024 = 9582.69             # €
ca_2025 = 17362.68            # €
ca_2025_corrige = 18625.96    # €
revpar_2025 = 51.07           # €
revpar_2025_corrige = 54.78   # €

# =====================
# HYPOTHÈSES
# =====================

jours_dispo = 340             # hypothèse : 25j maintenance/usage propriétaires (à valider)

# Scénario réaliste LMNP maintenu
tx_occ_s1 = 0.55
adr_s1 = 105.0
```

# Scénario réaliste LMNP maintenu

```python
ca_s1 = jours_dispo * tx_occ_s1 * adr_s1
revpar_s1 = tx_occ_s1 * adr_s1

print(f"CA estimé = {ca_s1:.2f} €")
print(f"RevPAR estimé = {revpar_s1:.2f} €")
```

```python
synthese = pd.DataFrame({
    'Indicateur': ['CA (€)', 'RevPAR (€)', 'Taux occupation', 'ADR (€)'],
    '2024 réel': [ca_2024, revpar_2024, f"{tx_occ_2024:.1%}", adr_2024],
    '2025 corrigé': [ca_2025_corrige, revpar_2025_corrige, 'N/A', 'N/A'],
    'Cible S1': [ca_s1, revpar_s1, f"{tx_occ_s1:.1%}", adr_s1]
})
synthese
```
